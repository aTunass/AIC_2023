```python
import os
import json
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
def get_youtube_video_id(youtube_url):
    # Tìm vị trí của "v=" trong URL
    start_index = youtube_url.find("v=")

    # Kiểm tra xem "v=" có tồn tại trong URL hay không
    if start_index != -1:
        # Tìm vị trí của ký tự "&" sau "v="
        end_index = youtube_url.find("&", start_index)

        # Nếu không tìm thấy ký tự "&", sử dụng độ dài của chuỗi làm chỉ số kết thúc
        if end_index == -1:
            video_id = youtube_url[start_index + 2:]
        else:
            video_id = youtube_url[start_index + 2:end_index]

        return video_id
    else:
        return None
```

```python
path = "/content/drive/MyDrive/metadata_edit2"
lst_dirs = os.listdir(path)
lst_dirs.sort()
data_asr = {}
lst_nosub = []
for lst_dir in lst_dirs:
  folder_path = os.path.join(path, lst_dir)
  print('get subtitle from', lst_dir[:-5])
  with open(folder_path, 'r') as file:
    data = json.load(file)
  video_id = get_youtube_video_id(data['watch_url'])
  print(video_id)
  try:
    subtitle = YouTubeTranscriptApi.get_transcript(video_id, languages=['vi', 'en'])
    data_asr[lst_dir[:-5]] = subtitle
  except:
    lst_nosub.append(lst_dir)
    pass
  # if lst_dir[:-5] == "L01_V030":
  #   break
```

```python
# Mở tệp văn bản để đọc
with open('/content/drive/MyDrive/AIC_2023/note title.txt', 'r') as file:
    # Sử dụng vòng lặp để đọc từng dòng trong tệp
    for line in file:
        print(line.strip())  # In dòng và loại bỏ ký tự xuống dòng (\n)

# Sau khi kết thúc vòng lặp, tệp sẽ tự động đóng
```

```python
result = {}  # Tạo một từ điển trống để lưu kết quả cuối cùng
current_key = None  # Biến để lưu key hiện tại trong tệp

# Đọc tệp văn bản
with open('/content/drive/MyDrive/AIC_2023/note title.txt', 'r') as file:
    for line in file:
        line = line.strip()  # Loại bỏ ký tự xuống dòng (newline)

        # Kiểm tra nếu dòng bắt đầu bằng "L", đây có thể là key
        if line.startswith("L"):
            current_key = line  # Cập nhật key hiện tại
            result[current_key] = []  # Tạo một danh sách rỗng cho key hiện tại
        else:
            # Phân tách dòng thành các phần
            parts = line.split(":")
            if len(parts) == 2:
                text = parts[0].strip()
                start, end = parts[1].strip().split("-")
                start, end = int(start), int(end)

                # Tạo một từ điển cho mỗi mục và thêm vào danh sách
                item = {"text": text, "start": start, "duration": end - start}
                result[current_key].append(item)

print(result)
```
